# Problem-10 Plot Stacked Bar chart quarter wise for each product.

```python
import pandas as pd 
import matplotlib.pyplot as plt 
a2=pd.read_csv(r"C:\Users\LENOVO\Downloads\company_sales_data - company_sales_data.csv")
a2["exact"]=pd.to_datetime([f"2020-{mn}-01" for mn in a2["month_number"]]) 
a3=a2.groupby(a2['exact'].dt.quarter).sum().iloc[:,1:7]
aq=[]
for mn in a3.columns:
    if len(aq)==0:
        plt.bar(a3.index,a3[mn],label=mn,width=0.06) 
    else:
        plt.bar(a3.index,a3[mn],bottom=sum(aq),width=0.06,label=mn)
    aq.append(a3[mn])
plt.legend(loc="lower center")
plt.xticks(a3.index)
plt.xlabel("Quarter") 
plt.ylabel("Products") 
plt.show() 
```
